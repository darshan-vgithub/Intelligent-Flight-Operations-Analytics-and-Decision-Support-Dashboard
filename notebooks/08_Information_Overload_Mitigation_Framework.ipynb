```python
# ============================================================
# Step 5 : Operational Context Extraction
# ============================================================

def extract_context(row):

    context = []

    # ---------------------------------------------------
    # Flight Phase
    # ---------------------------------------------------

    if row["flight_phase"] == "Ground":
        context.append("Aircraft On Ground")

    elif row["flight_phase"] == "Departure":
        context.append("Aircraft Departing")

    elif row["flight_phase"] == "Approach":
        context.append("Aircraft Approaching Airport")

    elif row["flight_phase"] == "Cruise":
        context.append("Aircraft in Cruise")

    # ---------------------------------------------------
    # Operational Risk
    # ---------------------------------------------------

    if row["operational_risk"] == "High":
        context.append("High Operational Risk")

    elif row["operational_risk"] == "Medium":
        context.append("Medium Operational Risk")

    else:
        context.append("Low Operational Risk")

    # ---------------------------------------------------
    # Airport Importance
    # ---------------------------------------------------

    if row["nearest_airport_importance"] == "High":
        context.append("Near Major Airport")

    elif row["nearest_airport_importance"] == "Medium":
        context.append("Near Regional Airport")

    else:
        context.append("Near Local Airport")

    # ---------------------------------------------------
    # Airport Size
    # ---------------------------------------------------

    if row["nearest_airport_size"] == "Large":
        context.append("Large Airport")

    elif row["nearest_airport_size"] == "Medium":
        context.append("Medium Airport")

    else:
        context.append("Small Airport")

    # ---------------------------------------------------
    # Aircraft Size
    # ---------------------------------------------------

    if row["Aircraft_Size_Category"] == "Wide-Body":
        context.append("Wide-Body Aircraft")

    elif row["Aircraft_Size_Category"] == "Narrow-Body":
        context.append("Narrow-Body Aircraft")

    elif row["Aircraft_Size_Category"] == "Regional Aircraft":
        context.append("Regional Aircraft")

    else:
        context.append("Small Aircraft")

    return ", ".join(context)


# Apply Context Extraction

df["Operational_Context"] = df.apply(extract_context, axis=1)

print("Operational Context Created Successfully!")

df[
    [
        "callsign",
        "flight_phase",
        "operational_risk",
        "nearest_airport_importance",
        "Aircraft_Size_Category",
        "Operational_Context"
    ]
].head()
```

```python
# ============================================================
# Step 6 : Priority Feature Selection
# ============================================================

priority_features = [

    "flight_phase",

    "operational_risk",

    "Aircraft_Size_Category",

    "Weight_Category",

    "nearest_airport_importance",

    "nearest_airport_size",

    "distance_to_airport_km"

]

print("="*60)
print("Priority Features Used")
print("="*60)

for feature in priority_features:
    print(feature)

print("\nTotal Priority Features :", len(priority_features))
```

```python
# ============================================================
# Step 7 : Rule-Based Priority Score
# ============================================================

def calculate_priority_score(row):

    score = 0

    # ---------------------------------------------
    # Operational Risk
    # ---------------------------------------------

    if row["operational_risk"] == "High":
        score += 40

    elif row["operational_risk"] == "Medium":
        score += 20

    else:
        score += 10

    # ---------------------------------------------
    # Flight Phase
    # ---------------------------------------------

    if row["flight_phase"] == "Approach":
        score += 20

    elif row["flight_phase"] == "Departure":
        score += 15

    elif row["flight_phase"] == "Cruise":
        score += 10

    elif row["flight_phase"] == "Ground":
        score += 5

    # ---------------------------------------------
    # Airport Importance
    # ---------------------------------------------

    if row["nearest_airport_importance"] == "High":
        score += 15

    elif row["nearest_airport_importance"] == "Medium":
        score += 10

    else:
        score += 5

    # ---------------------------------------------
    # Airport Distance
    # ---------------------------------------------

    if row["distance_to_airport_km"] <= 10:
        score += 15

    elif row["distance_to_airport_km"] <= 30:
        score += 10

    else:
        score += 5

    # ---------------------------------------------
    # Aircraft Size
    # ---------------------------------------------

    if row["Aircraft_Size_Category"] == "Wide-Body":
        score += 15

    elif row["Aircraft_Size_Category"] == "Narrow-Body":
        score += 10

    elif row["Aircraft_Size_Category"] == "Regional Aircraft":
        score += 8

    else:
        score += 5

    # ---------------------------------------------
    # Weight Category
    # ---------------------------------------------

    if row["Weight_Category"] == "Heavy":
        score += 10

    elif row["Weight_Category"] == "Medium":
        score += 7

    else:
        score += 4

    return score


df["Priority_Score"] = df.apply(calculate_priority_score, axis=1)

print("Priority Scores Generated Successfully!")

df[["callsign","Priority_Score"]].head()
```

```python
# ============================================================
# Step 8 : Alert Categorization
# ============================================================

def assign_alert(score):

    if score >= 90:
        return "Critical"

    elif score >= 70:
        return "High"

    elif score >= 50:
        return "Medium"

    else:
        return "Low"


df["Alert_Category"] = df["Priority_Score"].apply(assign_alert)

print("Alert Categories Created Successfully!")

print(df["Alert_Category"].value_counts())
```

```python
# =====================================================
# Step 9 : Alert Ranking
# =====================================================

df = df.sort_values(
    by="Priority_Score",
    ascending=False
)

df["Alert_Rank"] = range(1, len(df)+1)

print(df[
    [
        "callsign",
        "Priority_Score",
        "Alert_Category",
        "Alert_Rank"
    ]
].head(20))
```

```python
# =====================================================
# Step 10 : Information Reduction
# =====================================================

priority_flights = df[
    df["Alert_Category"].isin(
        [
            "Critical",
            "High"
        ]
    )
]

print("Original Flights :", len(df))

print("Priority Flights :", len(priority_flights))
```
