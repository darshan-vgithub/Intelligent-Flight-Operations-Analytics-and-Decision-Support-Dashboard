```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../../data/raw/delays/flights_sample_3m.csv")

print(df.shape)
```

```python
df_clean = df.copy()
```

```python
df_clean["FL_DATE"] = pd.to_datetime(df_clean["FL_DATE"])
```

```python
df_clean["FL_DATE"].dtype
```

```python
delay_columns = [
    "DELAY_DUE_CARRIER",
    "DELAY_DUE_WEATHER",
    "DELAY_DUE_NAS",
    "DELAY_DUE_SECURITY",
    "DELAY_DUE_LATE_AIRCRAFT"
]

df_clean[delay_columns] = df_clean[delay_columns].fillna(0)
```

```python
median_elapsed = df_clean["CRS_ELAPSED_TIME"].median()

df_clean["CRS_ELAPSED_TIME"] = df_clean["CRS_ELAPSED_TIME"].fillna(median_elapsed)
```

```python
missing = df_clean.isnull().sum()

missing = missing[missing > 0].sort_values(ascending=False)

missing
```

```python
print("Duplicate Records:", df_clean.duplicated().sum())
```

```python
df_clean.dtypes
```

```python
df_clean.to_csv(
    "../../data/processed/delays/delays_processed.csv",
    index=False
)

print("Cleaned dataset saved successfully.")
```
