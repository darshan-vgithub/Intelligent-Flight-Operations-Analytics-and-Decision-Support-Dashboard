```python
import pandas as pd
aircraft = pd.read_csv(
    "../../data/processed/aircraft/aircraft_metadata_processed.csv",
    parse_dates=[
        "LAST ACTION DATE",
        "CERT ISSUE DATE",
        "AIR WORTH DATE",
        "EXPIRATION DATE"
    ]
)

acftref = pd.read_csv(
    "../../data/raw/aircraft/ACFTREF.txt",
    low_memory=False
)

print(acftref.shape)

acftref.head()
```

```python
acftref.info()
```

```python
acftref.columns
```

```python
acftref = acftref.drop(columns=["Unnamed: 13"])
```

```python
print("Duplicate Codes:", acftref["CODE"].duplicated().sum())
```

```python
print(aircraft["MFR MDL CODE"].dtype)

print(acftref["CODE"].dtype)
```

```python
aircraft_metadata = aircraft.merge(
    acftref,
    left_on="MFR MDL CODE",
    right_on="CODE",
    how="left"
)
```

```python
print("="*60)
print("Merged Aircraft Dataset")
print("="*60)

print(aircraft_metadata.shape)
```

```python
aircraft_metadata.head()
```

```python
aircraft_metadata["MFR"].isnull().sum()
```

```python
aircraft_metadata["MODEL"].isnull().sum()
```

```python
matched = aircraft_metadata["MFR"].notna().sum()

total = len(aircraft_metadata)

print(f"Matched Records : {matched:,}")

print(f"Total Records   : {total:,}")

print(f"Match Rate      : {(matched/total)*100:.2f}%")
```

```python
aircraft_metadata.to_csv(
    "../../data/processed/aircraft/aircraft_metadata_enriched.csv",
    index=False
)

print("Aircraft Metadata Enriched Dataset Saved Successfully")
```

```python
aircraft_metadata.drop(
    columns=[
        "CODE",
        "Unnamed: 13"
    ],
    inplace=True
)
```

```python
aircraft_metadata.to_csv(
    "../../data/processed/aircraft/aircraft_metadata_final.csv",
    index=False
)

print("Final Aircraft Metadata Dataset Saved Successfully")
```
