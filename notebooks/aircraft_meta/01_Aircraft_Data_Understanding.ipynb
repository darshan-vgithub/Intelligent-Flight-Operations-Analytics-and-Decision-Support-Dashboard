```python
# ============================================================
# Missing Value Analysis
# ============================================================

missing_values = master.isnull().sum()

missing_percentage = (missing_values / len(master)) * 100

missing_summary = pd.DataFrame({
    "Missing Values": missing_values,
    "Percentage (%)": missing_percentage.round(2)
})

missing_summary = missing_summary.sort_values(
    by="Missing Values",
    ascending=False
)

missing_summary
```

```python
missing_summary[missing_summary["Missing Values"] > 0]
```

```python
plt.figure(figsize=(12,6))

missing_summary[missing_summary["Missing Values"] > 0]["Percentage (%)"].plot(
    kind="bar"
)

plt.title("Missing Values Percentage by Column")
plt.ylabel("Percentage (%)")
plt.xlabel("Columns")

plt.xticks(rotation=75)

plt.tight_layout()

plt.show()
```

```python
# ============================================================
# Empty String Analysis
# ============================================================

empty_strings = (master == "").sum()

empty_summary = pd.DataFrame({
    "Empty Strings": empty_strings
})

empty_summary = empty_summary.sort_values(
    by="Empty Strings",
    ascending=False
)

empty_summary.head(20)
```

```python
# ============================================================
# Detect Blank / Whitespace Values
# ============================================================

blank_summary = {}

for col in master.columns:
    if master[col].dtype == "object":
        blank_summary[col] = master[col].astype(str).str.strip().eq("").sum()

blank_summary = (
    pd.DataFrame.from_dict(
        blank_summary,
        orient="index",
        columns=["Blank Values"]
    )
    .sort_values(by="Blank Values", ascending=False)
)

blank_summary[blank_summary["Blank Values"] > 0]
```
