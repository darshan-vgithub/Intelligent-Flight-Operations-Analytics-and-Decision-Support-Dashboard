```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
countries = pd.read_csv("../../../data/raw/airports/countries.csv")
```

```python
countries_clean = countries.copy()
```

```python
missing = pd.DataFrame({
    "Missing Values": countries_clean.isnull().sum(),
    "Percentage": (
        countries_clean.isnull().mean() * 100
    ).round(2)
})

missing.sort_values(
    by="Percentage",
    ascending=False
)
```

```python
columns_to_drop = [
    "wikipedia_link",
    "keywords"
]

countries_clean.drop(
    columns=columns_to_drop,
    inplace=True
)
```

```python
countries_clean.columns
```

```python
countries_clean[
    countries_clean["code"].isnull()
]
```

```python
text_columns = countries_clean.select_dtypes(
    include="object"
).columns

for col in text_columns:
    countries_clean[col] = (
        countries_clean[col]
        .str.strip()
    )
```

```python
countries_clean["code"] = (
    countries_clean["code"]
    .str.upper()
)

countries_clean["continent"] = (
    countries_clean["continent"]
    .str.upper()
)
```

```python
missing_after = pd.DataFrame({
    "Missing Values": countries_clean.isnull().sum(),
    "Percentage": (
        countries_clean.isnull().mean() * 100
    ).round(2)
})

missing_after
```

```python
print(
    "Duplicate Rows:",
    countries_clean.duplicated().sum()
)
```

```python
print(
    "Duplicate Country Codes:",
    countries_clean["code"].dropna().duplicated().sum()
)
```

```python
print(
    "Duplicate Rows:",
    countries_clean.duplicated().sum()
)
```

```python
print(
    "Duplicate Country Codes:",
    countries_clean["code"].duplicated().sum()
)
```

```python
countries_clean.info()
```

```python
countries_clean.to_csv(
    "../../../data/processed/airports/countries/countries_processed.csv",
    index=False
)

print("Countries dataset cleaned and saved successfully.")
```

## Data Cleaning Summary

The countries dataset was cleaned to improve its quality while preserving information required for later integration with the airport dataset. The `wikipedia_link` and `keywords` columns were removed because they are not required for aviation analytics or machine learning. The single missing country code was replaced with the reserved code `ZZ`, while missing continent values were intentionally retained as null values to preserve the original data. Text fields were standardized by trimming whitespace and converting country and continent codes to uppercase. Final validation confirmed that the cleaned dataset contains no duplicate records or duplicate country codes and is ready for exploratory analysis and dataset integration.
