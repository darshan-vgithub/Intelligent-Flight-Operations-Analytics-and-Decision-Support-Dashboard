```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
regions = pd.read_csv("../../../data/raw/airports/regions.csv")
```

```python
regions_clean = regions.copy()
```

```python
missing = pd.DataFrame({
    "Missing Values": regions_clean.isnull().sum(),
    "Percentage": (
        regions_clean.isnull().mean() * 100
    ).round(2)
})

missing.sort_values(
    by="Percentage",
    ascending=False
)
```

```python
columns_to_drop = [
    "wikipedia_link",
    "keywords"
]

regions_clean.drop(
    columns=columns_to_drop,
    inplace=True
)
```

```python
text_columns = regions_clean.select_dtypes(
    include="object"
).columns

for col in text_columns:
    regions_clean[col] = (
        regions_clean[col]
        .str.strip()
    )
```

```python
regions_clean["code"] = (
    regions_clean["code"]
    .str.upper()
)

regions_clean["local_code"] = (
    regions_clean["local_code"]
    .str.upper()
)

regions_clean["continent"] = (
    regions_clean["continent"]
    .str.upper()
)

regions_clean["iso_country"] = (
    regions_clean["iso_country"]
    .str.upper()
)
```

```python
missing_after = pd.DataFrame({
    "Missing Values": regions_clean.isnull().sum(),
    "Percentage": (
        regions_clean.isnull().mean() * 100
    ).round(2)
})

missing_after
```

```python
print(
    "Duplicate Rows:",
    regions_clean.duplicated().sum()
)
```

```python
print(
    "Duplicate Region Codes:",
    regions_clean["code"].duplicated().sum()
)
```

```python
regions_clean.info()
```

```python
regions_clean.to_csv(
    "../../../data/processed/airports/regions/regions_processed.csv",
    index=False
)

print("Regions dataset cleaned and saved successfully.")
```

## Data Cleaning Summary

The regions dataset was cleaned by removing non-essential reference columns (`wikipedia_link` and `keywords`) while retaining all attributes required for geographic enrichment and future dataset integration. Text fields were standardized by trimming whitespace and converting region, country, and continent codes to uppercase. Missing values in `local_code`, `continent`, and `iso_country` were intentionally preserved because this dataset serves as a reference table, and modifying missing values could reduce data integrity. Final validation confirmed that the cleaned dataset contains no duplicate records or duplicate region codes and is suitable for exploratory analysis and later integration with the airport dataset.
