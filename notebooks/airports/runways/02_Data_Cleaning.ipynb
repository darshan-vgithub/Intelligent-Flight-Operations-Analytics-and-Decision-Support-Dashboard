```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
runways = pd.read_csv("../../../data/raw/airports/runways.csv")
```

```python
runways_clean = runways.copy()
```

```python
missing = pd.DataFrame({
    "Missing Values": runways_clean.isnull().sum(),
    "Percentage": (
        runways_clean.isnull().mean() * 100
    ).round(2)
})

missing.sort_values(
    by="Percentage",
    ascending=False
)
```

```python
runways_clean["length_ft"] = runways_clean["length_ft"].fillna(
    runways_clean["length_ft"].median()
)
```

```python
runways_clean["width_ft"] = runways_clean["width_ft"].fillna(
    runways_clean["width_ft"].median()
)
```

```python
runways_clean["surface"] = (
    runways_clean["surface"]
    .fillna("UNKNOWN")
)
```

```python
runways_clean["le_ident"] = (
    runways_clean["le_ident"]
    .fillna("UNKNOWN")
)

runways_clean["he_ident"] = (
    runways_clean["he_ident"]
    .fillna("UNKNOWN")
)
```

```python
text_columns = runways_clean.select_dtypes(
    include="object"
).columns

for col in text_columns:
    runways_clean[col] = (
        runways_clean[col]
        .str.strip()
        .str.upper()
    )
```

```python
missing_after = pd.DataFrame({
    "Missing Values": runways_clean.isnull().sum(),
    "Percentage": (
        runways_clean.isnull().mean() * 100
    ).round(2)
})

missing_after.sort_values(
    by="Percentage",
    ascending=False
)
```

```python
print(
    "Duplicate Rows:",
    runways_clean.duplicated().sum()
)
```

```python
print(
    "Duplicate Runway IDs:",
    runways_clean["id"].duplicated().sum()
)
```

```python
runways_clean.info()
```

```python
runways_clean.to_csv(
    "../../../data/processed/airports/runways/runways_processed.csv",
    index=False
)

print("Runways dataset cleaned and saved successfully.")
```

## Data Cleaning Summary

The runway dataset was cleaned by addressing missing values in key operational attributes while preserving detailed runway-end information for future analysis. Missing runway lengths and widths were imputed using the median values, while missing surface types and runway identifiers were replaced with the value `UNKNOWN`. Text fields were standardized by removing leading and trailing whitespace and converting categorical values to uppercase. Highly incomplete runway-end attributes were intentionally retained because they may provide additional operational information during later stages of feature engineering or future extensions of the project. Final validation confirmed that the cleaned dataset contains no duplicate records or duplicate runway identifiers and is suitable for exploratory data analysis.
