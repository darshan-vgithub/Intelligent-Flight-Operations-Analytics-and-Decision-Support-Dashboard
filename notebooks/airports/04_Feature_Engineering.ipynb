```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
airports = pd.read_csv(
    "../../data/processed/airports/airports_processed.csv"
)
```

```python
size_mapping = {
    "small_airport": "Small",
    "medium_airport": "Medium",
    "large_airport": "Large",
    "heliport": "Heliport",
    "seaplane_base": "Seaplane",
    "balloonport": "Balloon",
    "closed": "Closed"
}

airports["airport_size"] = airports["type"].map(size_mapping)
```

```python
airports["airport_size"].value_counts()
```

```python
def airport_category(t):

    if t == "large_airport":
        return "Major"

    elif t == "medium_airport":
        return "Regional"

    elif t == "small_airport":
        return "Local"

    elif t == "heliport":
        return "Heliport"

    else:
        return "Other"

airports["airport_category"] = airports["type"].apply(
    airport_category
)
```

```python
airports["commercial_airport"] = (
    airports["scheduled_service"]
    .map({
        "yes": 1,
        "no": 0
    })
)
```

```python
airports["commercial_airport"].value_counts()
```

```python
def elevation_band(elevation):

    if elevation < 1000:
        return "Low"

    elif elevation < 5000:
        return "Medium"

    elif elevation < 10000:
        return "High"

    else:
        return "Extreme"

airports["elevation_category"] = (
    airports["elevation_ft"]
    .apply(elevation_band)
)
```

```python
def hemisphere(latitude):
    if latitude >= 0:
        return "Northern"
    else:
        return "Southern"

airports["hemisphere"] = airports["latitude_deg"].apply(hemisphere)
```

```python
airports["hemisphere"].value_counts()
```

```python
def climate_zone(latitude):

    latitude = abs(latitude)

    if latitude < 23.5:
        return "Tropical"

    elif latitude < 40:
        return "Subtropical"

    elif latitude < 60:
        return "Temperate"

    else:
        return "Polar"

airports["climate_zone"] = airports["latitude_deg"].apply(climate_zone)
```

```python
airports["climate_zone"].value_counts()
```

```python
def airport_importance(row):

    if row["type"] == "large_airport" and row["commercial_airport"] == 1:
        return "High"

    elif row["type"] == "medium_airport" and row["commercial_airport"] == 1:
        return "Medium"

    else:
        return "Low"

airports["airport_importance"] = airports.apply(
    airport_importance,
    axis=1
)
```

```python
airports["airport_importance"].value_counts()
```

```python
engineered_features = [
    "airport_size",
    "airport_category",
    "commercial_airport",
    "elevation_category",
    "hemisphere",
    "climate_zone",
    "airport_importance"
]

airports[engineered_features].head()
```

```python
airports.to_csv(
    "../../data/processed/airports/airports_feature_engineered.csv",
    index=False
)

print("Airport feature engineered dataset saved successfully.")
```
