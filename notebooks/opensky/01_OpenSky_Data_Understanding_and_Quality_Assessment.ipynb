## 2.2 Missing Value Analysis

Missing values are a common issue in real-world aviation datasets due to factors such as sensor failures, temporary communication loss, incomplete ADS-B transmissions, or limitations in surveillance coverage. Identifying missing values is an essential step in assessing data quality, as they may influence statistical analysis, visualization, and machine learning model performance.

This section evaluates the completeness of each attribute by calculating the number and percentage of missing values in the OpenSky ADS-B dataset.

```python
# Count missing values in each column
missing_values = df.isnull().sum()

# Calculate percentage of missing values
missing_percentage = (missing_values / len(df)) * 100

# Create summary table
missing_summary = pd.DataFrame({
    "Missing Values": missing_values,
    "Percentage (%)": missing_percentage.round(2)
})

# Sort in descending order
missing_summary = missing_summary.sort_values(
    by="Missing Values",
    ascending=False
)

display(missing_summary)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

missing_summary["Missing Values"].plot(kind="bar")

plt.title("Missing Values in OpenSky ADS-B Dataset")
plt.xlabel("Attributes")
plt.ylabel("Number of Missing Values")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
completeness = pd.DataFrame({
    "Total Records": len(df),
    "Available": df.notnull().sum(),
    "Missing": df.isnull().sum()
})

completeness["Completeness (%)"] = (
    completeness["Available"] / len(df) * 100
).round(2)

display(completeness)
```

## 2.3 Duplicate Record Analysis

Duplicate records occur when two or more rows contain identical information across all attributes. In aviation datasets, duplicate records can arise from repeated data ingestion, synchronization issues, or processing errors. Such duplicates may bias statistical analyses and machine learning models by over-representing specific observations.

This section examines the dataset for exact duplicate rows.

```python
duplicate_count = df.duplicated().sum()

print("="*10)
print("Duplicate Record Analysis")
print("="*10)

print(f"Total Duplicate Records : {duplicate_count}")

if duplicate_count > 0:
    print("\nDuplicate Rows:")
    display(df[df.duplicated()])
else:
    print("\n✅ No duplicate records were found in the dataset.")
```

## 2.5 Unique Aircraft Analysis

The `icao24` attribute uniquely identifies an aircraft rather than an individual flight observation. Since each aircraft may appear multiple times within the dataset, repeated `icao24` values are expected. This analysis determines the number of unique aircraft represented in the dataset and identifies the most frequently observed aircraft.

```python
unique_aircraft = df["icao24"].nunique()

print("="*10)
print("Unique Aircraft Analysis")
print("="*10)

print(f"Total Records          : {len(df)}")
print(f"Unique Aircraft        : {unique_aircraft}")
print(f"Repeated Observations  : {len(df) - unique_aircraft}")

print("\nTop 10 Most Frequently Observed Aircraft")

display(
    df["icao24"]
    .value_counts()
    .head(10)
)
```

## 2.6 Flight Callsign Analysis

Unlike the `icao24` identifier, the `callsign` represents the flight currently being operated. A callsign is expected to appear in multiple observations while the aircraft is in flight. This analysis examines the number of unique flight callsigns and identifies the most frequently occurring callsigns.

```python
unique_callsigns = df["callsign"].nunique()

print("="*10)
print("Flight Callsign Analysis")
print("="*10)

print(f"Unique Callsigns : {unique_callsigns}")

print("\nTop 10 Most Frequent Callsigns")

display(
    df["callsign"]
    .value_counts()
    .head(10)
)
```

## 2.6 Geographic Coordinate Validation

Aircraft positions are represented using latitude and longitude coordinates. These values must lie within valid geographical ranges to ensure accurate flight tracking, route analysis, and map-based visualizations. This section validates the geographical coordinates recorded in the OpenSky ADS-B dataset.

```python
print("="*10)
print("Latitude Range")
print("="*10)

print(f"Minimum Latitude : {df['latitude'].min()}")
print(f"Maximum Latitude : {df['latitude'].max()}")

print()

print("="*10)
print("Longitude Range")
print("="*10)

print(f"Minimum Longitude : {df['longitude'].min()}")
print(f"Maximum Longitude : {df['longitude'].max()}")
```

```python
invalid_latitude = df[
    (df["latitude"] < -90) |
    (df["latitude"] > 90)
]

invalid_longitude = df[
    (df["longitude"] < -180) |
    (df["longitude"] > 180)
]

print(f"Invalid Latitude Records  : {len(invalid_latitude)}")
print(f"Invalid Longitude Records : {len(invalid_longitude)}")
```

```python
coordinate_summary = pd.DataFrame({
    "Attribute": ["Latitude", "Longitude"],
    "Minimum": [df["latitude"].min(), df["longitude"].min()],
    "Maximum": [df["latitude"].max(), df["longitude"].max()],
    "Missing": [df["latitude"].isnull().sum(), df["longitude"].isnull().sum()]
})

display(coordinate_summary)
```

```python
aviation_columns = [
    "velocity",
    "baro_altitude",
    "geo_altitude",
    "true_track",
    "vertical_rate"
]

aviation_summary = pd.DataFrame({
    "Minimum": df[aviation_columns].min(),
    "Maximum": df[aviation_columns].max(),
    "Mean": df[aviation_columns].mean(),
    "Missing": df[aviation_columns].isnull().sum()
})

display(aviation_summary)
```

```python
invalid_heading = df[
    (df["true_track"] < 0) |
    (df["true_track"] > 360)
]

print("="*10)
print("Heading Validation")
print("="*10)

print(f"Invalid Heading Records : {len(invalid_heading)}")
```

```python
negative_baro = df[df["baro_altitude"] < 0]
negative_geo = df[df["geo_altitude"] < 0]

print("="*10)
print("Negative Altitude Check")
print("="*10)

print(f"Negative Barometric Altitude : {len(negative_baro)}")
print(f"Negative Geometric Altitude  : {len(negative_geo)}")
```
