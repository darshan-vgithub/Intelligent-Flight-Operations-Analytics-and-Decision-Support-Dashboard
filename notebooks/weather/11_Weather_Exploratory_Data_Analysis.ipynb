```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

plt.style.use("ggplot")

print("Libraries Imported Successfully!")
```

```python
weather = pd.read_csv(
    "../../data/processed/weather/weather_processed.csv"
)

weather["datetime"] = pd.to_datetime(weather["datetime"])

print("Dataset Loaded Successfully!")
```

```python
print(weather.shape)

weather.head()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    weather["temperature"],
    bins=40,
    kde=True
)

plt.title("Temperature Distribution")

plt.xlabel("Temperature (°C)")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    weather["humidity"],
    bins=40,
    kde=True
)

plt.title("Humidity Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    weather["pressure"],
    bins=40,
    kde=True
)

plt.title("Pressure Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    weather["wind_direction"],
    bins=36
)

plt.title("Wind Direction Distribution")

plt.show()
```

```python
plt.figure(figsize=(14,8))

weather["weather_description"]\
.value_counts()\
.head(15)\
.plot(kind="bar")

plt.title("Top Weather Conditions")

plt.show()
```

```python
plt.figure(figsize=(12,6))

weather["city"]\
.value_counts()\
.plot(kind="bar")

plt.title("Observations by City")

plt.show()
```

```python
city_temp = (
    weather
    .groupby("city")["temperature"]
    .mean()
    .sort_values()
)

plt.figure(figsize=(12,8))

city_temp.plot(kind="barh")

plt.title("Average Temperature by City")

plt.show()
```

```python
city_humidity = (
    weather
    .groupby("city")["humidity"]
    .mean()
    .sort_values()
)

plt.figure(figsize=(12,8))

city_humidity.plot(kind="barh")

plt.title("Average Humidity by City")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    weather[
        [
            "temperature",
            "humidity",
            "pressure",
            "wind_direction"
        ]
    ].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Weather Feature Correlation")

plt.show()
```

```python
weather["month"] = weather["datetime"].dt.month

monthly = (
    weather
    .groupby("month")["temperature"]
    .mean()
)

plt.figure(figsize=(10,6))

monthly.plot(marker="o")

plt.title("Average Monthly Temperature")

plt.xlabel("Month")

plt.ylabel("Temperature (°C)")

plt.show()
```

```python
monthly = (
    weather
    .groupby("month")["humidity"]
    .mean()
)

plt.figure(figsize=(10,6))

monthly.plot(marker="o")

plt.title("Average Monthly Humidity")

plt.show()
```

```python
monthly = (
    weather
    .groupby("month")["pressure"]
    .mean()
)

plt.figure(figsize=(10,6))

monthly.plot(marker="o")

plt.title("Average Monthly Pressure")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(x=weather["humidity"])

plt.title("Humidity Boxplot")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(x=weather["temperature"])

plt.title("Temperature Boxplot")

plt.show()
```
