```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

print("Libraries Imported Successfully")
```

```python
weather = pd.read_csv(
    "../../data/intermediate/weather/weather_master.csv"
)

print("Weather Dataset Loaded Successfully")
```

```python
print("Shape:", weather.shape)

weather.head()
```

```python
weather["datetime"] = pd.to_datetime(weather["datetime"])

print(weather["datetime"].dtype)
```

```python
missing = weather.isnull().sum().to_frame(name="Missing Values")

missing["Percentage"] = (
    missing["Missing Values"] / len(weather) * 100
).round(2)

missing
```

```python
weather = weather.sort_values(["city", "datetime"])
```

```python
weather["temperature"] = (
    weather.groupby("city")["temperature"]
    .transform(lambda x: x.interpolate())
)
```

```python
weather["humidity"] = (
    weather.groupby("city")["humidity"]
    .transform(lambda x: x.interpolate())
)
```

```python
weather["pressure"] = (
    weather.groupby("city")["pressure"]
    .transform(lambda x: x.interpolate())
)
```

```python
weather["weather_description"] = (
    weather.groupby("city")["weather_description"]
    .transform(lambda x: x.ffill().bfill())
)
```

```python
weather.isnull().sum()
```

```python
weather["temperature"] = weather["temperature"] - 273.15
```

```python
weather["temperature"].describe()
```

```python
weather[
    (weather["humidity"] < 0)
    | (weather["humidity"] > 100)
]
```

```python
weather[
    (weather["pressure"] < 800)
    | (weather["pressure"] > 1100)
]
```

```python
weather[
    (weather["wind_direction"] < 0)
    | (weather["wind_direction"] > 360)
]
```

```python
print(
    "Duplicate Records:",
    weather.duplicated().sum()
)
```

```python
weather.to_csv(
    "../../data/processed/weather/weather_processed.csv",
    index=False
)

print("Processed Weather Dataset Saved Successfully!")
```
