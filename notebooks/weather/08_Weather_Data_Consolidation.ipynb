```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

print("Libraries Imported Successfully")
```

```python
temperature = pd.read_csv("../../data/raw/weather/temperature.csv")
humidity = pd.read_csv("../../data/raw/weather/humidity.csv")
pressure = pd.read_csv("../../data/raw/weather/pressure.csv")
wind = pd.read_csv("../../data/raw/weather/wind_direction.csv")
description = pd.read_csv("../../data/raw/weather/weather_description.csv")

print("All datasets loaded successfully!")
```

```python
temperature_long = temperature.melt(
    id_vars="datetime",
    var_name="city",
    value_name="temperature"
)

temperature_long.head()
```

```python
humidity_long = humidity.melt(
    id_vars="datetime",
    var_name="city",
    value_name="humidity"
)
```

```python
pressure_long = pressure.melt(
    id_vars="datetime",
    var_name="city",
    value_name="pressure"
)
```

```python
wind_long = wind.melt(
    id_vars="datetime",
    var_name="city",
    value_name="wind_direction"
)
```

```python
description_long = description.melt(
    id_vars="datetime",
    var_name="city",
    value_name="weather_description"
)
```

```python
weather = temperature_long.merge(
    humidity_long,
    on=["datetime", "city"]
)

weather = weather.merge(
    pressure_long,
    on=["datetime", "city"]
)

weather = weather.merge(
    wind_long,
    on=["datetime", "city"]
)

weather = weather.merge(
    description_long,
    on=["datetime", "city"]
)

weather.head()
```

```python
print(weather.shape)
```

```python
weather.to_csv(
    "../../data/intermediate/weather/weather_master.csv",
    index=False
)

print("Weather Master Dataset Saved Successfully!")
```
